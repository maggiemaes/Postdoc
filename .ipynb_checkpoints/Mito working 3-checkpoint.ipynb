```python
df_pcnt = df_raw[df_raw.Variable == 'Volume']
df_pcnt = df_pcnt.groupby(['surface_type', 'sex','condition', 'retinal_layer', 'mod_sex', 'mod_cond', 'mod_retinal_layer']).sum()['Value']
df_pcnt = (df_pcnt.loc['cd68'] / df_pcnt.loc['mg'])*100      #total cd68 volume/ total mg volume per each cell
df_pcnt = df_pcnt.reset_index()
df_pcnt['surface_type'] = 'pcnt_cd68_volume' 

# df_pcnt.to_csv('/Users/mmaes/Documents/Python_scripts/Output/cd68pcnt.csv')
```

```python
df_test = df_raw[(df_raw.condition == 'od1') & (df_raw.retinal_layer == 'ipl') & (df_raw.sex == 'fem1')]

mitolyso = df_raw[df_raw.surface_type == 'mitolyso']
mitolyso = mitolyso.groupby(['Variable', 'sex','condition', 'retinal_layer', 'mod_sex', 'mod_cond', 'mod_retinal_layer']).sum()['Value']
    
mito_area = df_raw[(df_raw.surface_type == 'mito') & (df_raw.Variable == 'Area')]
mito_area = mito_area.groupby(['Variable', 'sex','condition', 'retinal_layer', 'mod_sex', 'mod_cond', 'mod_retinal_layer']).sum()['Value']
    
pcnt_mito = (mitolyso.loc['Contact surface area'] / mito_area.loc['Area'])*100
pcnt_mito = pcnt_mito.reset_index()
pcnt_mito['surface_type'] = ' pcnt mito contact Cd68'
x = (mitolyso.loc['Area']/ mito_area.loc['Area'])*100
x= x.reset_index()

# pcnt_mito.to_csv('/Users/mmaes/Documents/Python_scripts/Output/mitocontact_ContactSurfArea.csv')
```
